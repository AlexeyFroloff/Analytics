```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

#Подготовка данных

```python
df = pd.read_excel("Analytic_test.xlsx", header = 1, index_col = 0)
```

```python
df.head(10)
```

```python
red = df.iloc[:, [0, 1, 11, 15, 19, 20]]
green = df.iloc[:, [0, 2, 7, 8, 12, 16, 19, 21]]
black = df.iloc[:, [0, 3, 9, 10, 13, 17, 19, 22]]
blue = df.iloc[:, [0, 4, 5, 6, 14, 18, 19, 23]]

red = red[red["active_Red"] != 0]
green = green[green["has_Green_id"] != 0]
black = black[black["has_Black_id"] != 0]
blue = blue[blue["has_Blue_id"] != 0]
```

#Green

```python
green.head(5)
```

```python
green_msk = green[green["is_msk"]==0]
green_notmsk = green[green["is_msk"]!=0]
```

```python
plt.figure(figsize = (12, 8))
plt.bar(["Москва", "Не Москва"], [green_msk.shape[0], green_notmsk.shape[0]], color = ["r", "b"])
plt.title('GREEN. Распределение пользователей по городам')
```

Выведем соотношения пользователей по платформам:

```python
green_web = green[green["Green_web"]!=0]
green_andr = green[green["Green_android"]!=0]
green_appl = green[green["Green_apple"]!=0]
green_others = green[(green["Green_web"]==0) * (green["Green_android"]==0) * (green["Green_apple"]==0)]
```

```python
plt.figure(figsize = (12, 8))
plt.bar(["Web", "android", "apple", "others"], [green_web.shape[0], green_andr.shape[0], green_appl.shape[0], green_others.shape[0]], color = ["g", "g", "g", "g"])
plt.title('GREEN. Распределения пользователей по платформам')
```

```python
GRgreen = green.groupby(["Green_android", "Green_apple", "Green_web"]).mean("Green_gvm_total")
```

```python
MAXgreen = green.groupby(["Green_android", "Green_apple", "Green_web"]).max("Green_gvm_total")
MAXgreen
```

```python
GRgreen
```

```python
GRgreen = green.groupby(["Green_android", "Green_apple", "Green_web", "is_msk"])["Green_gvm_total"].mean("Green_gvm_total")
COgreen = green.groupby(["Green_android", "Green_apple", "Green_web", "is_msk"])["HID"].count()
GREEN = pd.DataFrame(GRgreen).join(COgreen)
GREEN.columns = ["mean_gvm", "count"]
GREEN
```

Разобьем по 4 признакам юзеров на группы, и посмотрим на gmv. Визуализировать эту информацию трудно, но можно посмотреть на цифры в таблице и сделать выводы. Мы видим огромные цифры, но с ними надо быть аккуратнее: это скорее всего выбросы, обусловленные маленькой выборкой. Если смотреть на кластеры, в которых относительно большое количество пользователей, получим следующий вывод: наибольший оборот (GVM) у apple-юзеров из Москвы, относительно средний у android-юзеров. Поэтому компании Green следует проводить рекламу, ориентируясь на эти аудитории.
